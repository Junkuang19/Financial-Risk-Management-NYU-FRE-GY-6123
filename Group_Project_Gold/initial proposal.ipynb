## 2) Proposed Claim / Hypotheses / Arguments & Methodology

### 2.1 Thesis (one sentence)

**Gold’s unconditional buy-and-hold is modest, but a risk-managed approach that is *real-rate–aware* and *term-structure–informed* can materially improve risk-adjusted returns while preserving gold’s diversification during equity stress and inflation surprises.**

### 2.2 Research Questions & Hypotheses

* **RQ1 – Macro drivers.** What explains time-variation in gold returns?
  **H1:** **Real yields** (TIPS) and the **broad USD** dominate; **inflation surprises** and **risk-off proxies** (VIX/MOVE) add state dependence.
* **RQ2 – Portfolio role.** How effective is gold as a hedge/safe haven?
  **H2:** Gold has **positive convex exposure** to inflation surprises and **negative beta to rising real rates**; safe-haven behavior strengthens in equity drawdowns.
* **RQ3 – Implementable strategies.** Can simple rules beat passive exposure after costs?
  **H3:** A **curve-carry long/flat** rule with a **real-rate overlay** and **10–12% volatility targeting** outperforms buy-and-hold on Sharpe and max drawdown.

### 2.3 Methodology

**Data Construction**

* **Prices/Returns:** LBMA spot (London PM), COMEX front/next futures to build **constant-maturity** 1- and 3-month series; GLD NAV/flows (shares outstanding).
* **Macro factors:** 10y TIPS real yield (DFII10), 3m T-bill, CPI (monthly) with **surprise** = actual − survey/nowcast, USD broad index (DTWEXBGS/DXY proxy), VIX, MOVE.
* **Positioning/Flows (robustness):** CFTC COT for gold; GLD flows.

**Return Decomposition (commodity lens)**

* Decompose futures total return into **spot return + roll/carry (curve) + collateral yield (Rf)**.
* Classify **backwardation/contango** regimes; estimate average roll contribution by regime.

**Empirical Tests**

1. **Time-series regressions** of gold excess returns on Δreal yields, ΔUSD, CPI surprise, ΔVIX/ΔMOVE; rolling betas for regime shifts.
2. **Event studies** around CPI/FOMC/UST shocks (±5/10 trading days).
3. **Hedge & safe-haven tests:** conditional betas vs S&P 500 and a 60/40; performance in worst equity months (bottom decile).
4. **Strategies (out-of-sample discipline):**

   * **Curve-carry rule:** Long when carry > threshold/backwardation; otherwise flat/reduced weight.
   * **Real-rate overlay:** Scale exposure by –z(Δreal yield, 1–3m).
   * **Vol targeting:** Target 10–12% ann. vol; monthly rebalance; include trading costs and realistic rolls.
   * **Risk controls:** turnover caps, position bounds, optional drawdown guard.
5. **Evaluation metrics:** Ann. return/vol, Sharpe, Sortino, Max DD, skew, **equity/bond betas**, crisis-period performance, turnover & cost impact, **Newey–West t-stats**, subperiod stability.

**Implementation standards**

* Continuous compounding for rate transforms; explicit roll conventions; reproducible notebooks; slippage/fee sensitivity; train/test split for signals.
