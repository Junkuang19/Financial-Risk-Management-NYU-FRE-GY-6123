#### Problem Interpretation

We are asked to price a European-style derivative that pays the square of the logarithm of the terminal stock price:

$$
\text{Payoff} = \left( \log S(T) \right)^2
$$

To find its fair value at time $t$, we apply the Black-Scholes framework and risk-neutral valuation.

#### Risk-Neutral Pricing Formula

Using the standard pricing formula under risk-neutral measure $\mathbb{Q}$, the time-$t$ price is:

$$
LL(S,t) = e^{-r(T - t)} \cdot \mathbb{E}^{\mathbb{Q}} \left[ \left( \log S(T) \right)^2 \,\middle|\, S(t) = S \right]
$$

#### Distribution of $\log S(T)$ under $\mathbb{Q}$

In the Black-Scholes model, under the risk-neutral measure, the asset price follows the SDE:

$$
\frac{dS(t)}{S(t)} = r\,dt + \sigma\,dW_t^{\mathbb{Q}}
$$

Solving this gives:

$$
S(T) = S(t) \cdot \exp\left( \left(r - \frac{1}{2}\sigma^2\right)(T - t) + \sigma \sqrt{T - t} \cdot Z \right), \quad Z \sim \mathcal{N}(0,1)
$$

Taking logs:

$$
\log S(T) = \log S(t) + \left(r - \frac{1}{2}\sigma^2\right)(T - t) + \sigma \sqrt{T - t} \cdot Z
$$

So $\log S(T)$ is normally distributed with:

Mean:

$$
\mu = \log S(t) + \left(r - \frac{1}{2}\sigma^2\right)(T - t)
$$

Variance:

$$
\sigma^2 (T - t)
$$

#### Compute the Expectation

We now compute the expectation of the squared log:

$$
\mathbb{E}^{\mathbb{Q}}\left[ \left( \log S(T) \right)^2 \right]
$$

Using the identity for a normal variable:

$$
\mathbb{E}[X^2] = \mathrm{Var}(X) + \left(\mathbb{E}[X]\right)^2
$$

We get:

$$
\mathbb{E}^{\mathbb{Q}}\left[ \left( \log S(T) \right)^2 \right]
= \sigma^2 (T - t) + \left( \log S(t) + \left( r - \frac{1}{2} \sigma^2 \right)(T - t) \right)^2
$$

#### (b) Compute Delta and Gamma

Recall from part (a), the pricing function is:

$$
LL(S,t) = e^{-r(T-t)} \left[
\sigma^2 (T-t) + \left( \log S + \left(r - \tfrac{1}{2}\sigma^2\right)(T-t) \right)^2
\right]
$$

To compute delta, differentiate with respect to $S$:

$$
\frac{\partial LL}{\partial S}
= e^{-r(T-t)} \cdot 2\left( \log S + \left(r - \tfrac{1}{2}\sigma^2\right)(T-t) \right)\cdot \frac{1}{S}
$$

So the delta is:

$$
\boxed{
\Delta = \frac{2\,e^{-r(T-t)}}{S}
\left( \log S + \left(r - \tfrac{1}{2}\sigma^2\right)(T-t) \right)
}
$$

To compute gamma, differentiate again:

$$
\Gamma = \frac{\partial^2 LL}{\partial S^2}
= e^{-r(T-t)} \cdot \left[
\frac{-2}{S^2} \left( \log S + \left(r - \tfrac{1}{2}\sigma^2\right)(T-t) \right) + \frac{2}{S^2}
\right]
$$

Simplify:

$$
\boxed{
\Gamma = \frac{2e^{-r(T-t)}}{S^2}
\left[1 - \log S - \left(r - \tfrac{1}{2}\sigma^2\right)(T-t)\right]
}
$$

From part (a), note that $LL(S,t)$ includes a direct term in $\sigma^2(T - t)$ and an indirect dependence via:

#### (d) What type of leverage does this derivative provide?

To understand leverage, consider the **percentage change in the derivative’s value** per **percentage change in the underlying price**:

$$
\frac{\Delta LL / LL}{\Delta S / S} \approx \frac{S \cdot \Delta}{LL}
$$

Substitute the expression for $\Delta$:

$$
\frac{S \cdot \Delta}{LL} = \frac{2\left(\log S + \left(r - \tfrac{1}{2}\sigma^2\right)(T - t)\right)}
{\sigma^2(T - t) + \left(\log S + \left(r - \tfrac{1}{2}\sigma^2\right)(T - t)\right)^2}
$$

This expression shows:

- When $\log S$ is large in magnitude, leverage tends to decrease due to the denominator growing faster;
- When $\log S$ is near zero (i.e., $S$ near 1 in normalized units), the ratio becomes large — implying high sensitivity.

This means:

$$
\boxed{
\text{The derivative provides convex, nonlinear leverage. Sensitivity to } S \text{ is highest when } \log S \text{ is moderate.}
}
$$

We are given the following SDE for a process $X(t)$:

$$
dX(t) = \alpha X(t)\,dt + \sigma(t)\,dW(t)
$$

where $\alpha$ is a constant, and $\sigma(t)$ is a general stochastic process.

#### **a:** Situations where this model applies

This model represents a **multiplicative drift** (due to $\alpha X(t)$) plus an **additive stochastic noise** with time-varying volatility. A good example would be:

- A risky asset whose expected return is proportional to its current level ($\alpha X$),  
- But whose **volatility** is driven by an external process, e.g., stochastic volatility or random environmental factors.

One concrete example is:

> **Electricity spot prices or weather-influenced assets**, where the price evolves exponentially over time but volatility is driven by temperature, wind, or other random processes (e.g., $\sigma(t)$ follows mean-reverting dynamics around a long-term level).

If we assume that $\sigma(t)$ fluctuates around a constant value $\bar{\sigma}$, then this model describes a system with **steady-state average volatility** and random short-term deviations — consistent with real-world observations in many financial or physical systems.

$$
\boxed{
\text{This model captures exponential growth + stochastic noise with dynamic but bounded volatility.}
}
$$

---

#### **b:** Role of $X(t)$ in $\sigma(t)$ and correlation

If the stochastic volatility term $\sigma(t)$ depends on $X(t)$, say:

$$
\sigma(t) = f(X(t))
$$

then $dX(t)$ becomes a function of $X(t)$ in both drift and diffusion terms:

$$
dX(t) = \alpha X(t)\,dt + f(X(t))\,dW(t)
$$

This creates a feedback loop: the size of the noise term depends on the current value of $X(t)$, and vice versa.

Now suppose there is **another process**, say $Y(t)$, modeling **volatility**, and both $X(t)$ and $Y(t)$ are driven by correlated Brownian motions:

$$
\begin{aligned}
dX(t) &= \alpha X(t)\,dt + \sigma(t)\,dW_1(t) \\
d\sigma(t) &= \mu(t)\,dt + \nu(t)\,dW_2(t) \\
\text{with } \mathbb{E}[dW_1(t)\,dW_2(t)] = \rho\,dt
\end{aligned}
$$

Then by allowing $\sigma(t)$ to depend on $X(t)$ and making $dW_1$ and $dW_2$ correlated, we can **control the instantaneous correlation** between changes in $X(t)$ and changes in volatility.

- If $\rho < 0$, this models **leverage effect**: when $X(t)$ drops, volatility tends to rise.
- If $\rho > 0$, we get **positive stock-vol correlation**, as in some commodity markets.

Thus:

$$
\boxed{
\text{By embedding } X(t) \text{ inside the SDE for } \sigma(t), \text{ we can control stock-vol correlation via } \rho.
}
$$

```python
# Problem 3 (a) - Simple Simulating Brownian Motion and Its Running Maximum

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
T = 1.0           # total time
N = 500           # number of time steps
alpha = 1.0       # scaling factor
dt = T / N
t = np.linspace(0, T, N + 1)

# Simulate Brownian motion
dW = np.random.normal(0, np.sqrt(dt), N)
W = np.concatenate(([0], np.cumsum(dW)))   # standard Brownian motion
X = alpha * W                              # scaled process
M = np.maximum.accumulate(X)               # running maximum process

# plot the results
plt.figure(figsize=(10, 5))
plt.plot(t, X, label="X(t): Brownian Path", color="dodgerblue", linewidth=1.5)
plt.plot(t, M, label="M(t): Running Maximum", color="crimson", linestyle="--", linewidth=2)
plt.axhline(0, color="gray", linestyle=":", linewidth=1)
plt.title("Brownian Motion X(t) and Its Running Maximum M(t)", fontsize=13)
plt.xlabel("Time t")
plt.ylabel("Value")
plt.legend()
plt.grid(True, linestyle=":", alpha=0.6)
plt.tight_layout()
plt.show()
```

#### Problem 3 Continued

We are given a scaled Brownian motion process:

$$
dX(t) = \alpha\, dW(t), \quad X(0) = 0
$$

Let $M(T)$ be the maximum of this process over $[0, T]$:

$$
M(T) = \max_{0 \le u \le T} X(u)
$$

We are interested in characterizing the distribution of $M(T)$.

#### (b) Decompose the tail probability of $M(T)$

For any $u > 0$:

$$
1 - G(u) = \mathbb{P}(M(T) \ge u)
= \mathbb{P}(M(T) \ge u \text{ and } X(T) \ge u)
+ \mathbb{P}(M(T) \ge u \text{ and } X(T) < u)
$$

This simply partitions the event that the process has ever exceeded $u$ into two disjoint sub-events depending on whether the endpoint $X(T)$ is above or below $u$.

#### (c) Use path reflection to equate the two terms

By the **reflection principle**, every path that hits level $u$ and ends below $u$ has a mirror image path that ends above $u$. These two sets of paths are equally likely.

Therefore,

$$
\mathbb{P}(M(T) \ge u \text{ and } X(T) < u)
=
\mathbb{P}(M(T) \ge u \text{ and } X(T) \ge u)
$$

#### (d) Combine the two parts

From above:

$$
1 - G(u)
= \mathbb{P}(M(T) \ge u)
= 2 \cdot \mathbb{P}(M(T) \ge u \text{ and } X(T) \ge u)
$$

So,

$$
\boxed{
1 - G(u) = 2 \cdot \mathbb{P}(M(T) \ge u,\, X(T) \ge u)
}
$$

#### (e) Relate to $X(T)$ distribution

Notice that if $X(T) \ge u$, then necessarily the maximum $M(T) \ge u$ (since the process ends above $u$). So:

$$
\mathbb{P}(M(T) \ge u \text{ and } X(T) \ge u)
= \mathbb{P}(X(T) \ge u)
$$

Hence:

$$
\boxed{
\mathbb{P}(M(T) \ge u,\, X(T) \ge u) = \mathbb{P}(X(T) \ge u)
}
$$

#### (f) Final expression for the distribution function $G(u)$

Substitute into earlier identity:

$$
1 - G(u) = 2 \cdot (1 - F(u))
$$

where $F(u)$ is the CDF of $X(T) \sim \mathcal{N}(0, \alpha^2 T)$. Differentiate both sides:

$$
g(u) = \frac{dG}{du} = 2 f(u)
$$

So the density function of $M(T)$ is **twice the PDF** of the terminal value.

$$
\boxed{
g(u) = 2 f(u) = \frac{2}{\sqrt{2\pi \alpha^2 T}} \exp\left( -\frac{u^2}{2\alpha^2 T} \right)
}
$$

#### (g) Density function of $M(T)$

From above, since $X(T) \sim \mathcal{N}(0, \alpha^2 T)$, we have:

$$
f(u) = \frac{1}{\sqrt{2\pi \alpha^2 T}} \exp\left( -\frac{u^2}{2\alpha^2 T} \right)
$$

So:

$$
\boxed{
g(u) = \frac{2}{\sqrt{2\pi \alpha^2 T}} \exp\left( -\frac{u^2}{2\alpha^2 T} \right), \quad u \ge 0
}
$$

This is the **PDF of the maximum** of scaled Brownian motion over $[0, T]$.

#### (h) Why $M(t)$ has no simple SDE

Unlike $X(t)$, the maximum process $M(t)$ is **non-Markovian** and **non-differentiable**. It depends on the entire path history of $X(u)$ up to time $t$:

- It increases only when $X(t)$ hits a new high;
- It stays flat otherwise.

This behavior is **path-dependent**, not governed by a fixed local rule like an SDE.

Thus, we cannot write a standard Itô SDE for $M(t)$:

$$
\boxed{
\bf{M(t)} \text{ is not governed by a conventional SDE due to its non-local, path-dependent definition.}
}
$$
