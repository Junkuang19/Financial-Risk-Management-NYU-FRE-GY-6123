## Objective

The goal is to design a contingent **New Deal** transaction that maximizes the total value of **Debt + Equity**, while ensuring that the new investor receives a fair valuation and earns a **present-value profit of $10**. All payments occur at time $T$, and the firm currently has no cash, so the New Deal determines the only inflows/outflows prior to debt maturity.

---

## 1. Baseline Model Without the New Deal

The firm follows a Merton-style capital structure with financial distress costs and corporate taxes:

* Asset value at maturity: $AT$, lognormally distributed
* Face value of debt: $F = 80$
* Distress loss: $ZT = \min(Z, AT)\mathbf{1}(AT < F)$
* Debt payoff: $DT = \min(AT - ZT, F)$
* Equity payoff: $ET = (1-\tau)(AT - DT - ZT)$
* Government taxes: $GT = \tau(AT - DT - ZT)$

Under this structure, low-asset states generate default and distress costs, reducing the combined value of debt and equity. High-asset states capture upside, but this upside does not fully offset value destroyed in default states.

---

## 2. Economic Logic Behind the New Deal

A value-maximizing New Deal must:

1. Reduce or eliminate distress costs in low-asset states (large benefit to firm value).
2. Provide upside participation to the new investor in high-asset states (so they can earn $+10$ PV).
3. Include positive inflows in some outcomes and negative outflows in others, consistent with the assignment instructions.

To achieve this, the New Deal is designed as a state-contingent transfer $NT(AT)$, which adjusts the firmâ€™s effective asset value to $AT + NT$ before applying the waterfall.

---

## 3. Proposed New Deal Structure

The New Deal is defined by a single parameter $\alpha$ that governs how much high-state upside is shared with the investor:

> For $AT < F$: $NT(AT) = F - AT$
>
> For $AT \ge F$: $NT(AT) = -\alpha (AT - F)$

### Interpretation

* **Low states ($AT < F$)**
  The investor injects exactly enough capital to lift the firmâ€™s asset value to $F$.
  â‡’ Default and distress costs $ZT$ drop to zero, and debt is fully repaid.

* **High states ($AT \ge F$)**
  The firm shares an $\alpha$-fraction of the upside above $F$ with the investor.
  â‡’ The investor captures meaningful upside, enabling them to earn a $10 PV profit.

The optimization chooses $\alpha$ to maximize firm value (Debt + Equity) subject to:

* Investor PV $= 10$
* And the sign relation: $\text{Investor PV} = - \text{NewDeal AveragePV}$ (from the firmâ€™s perspective).

---

## 4. Why This Structure Maximizes Value

### Value creation

* In low states, injections via $NT(AT) = F - AT$ eliminate distress losses $ZT$, which are highly value-destroying.
* The firm benefits significantly from removing default and expected bankruptcy costs.

### Why upside sharing is necessary

* Without upside participation, an investor who only injects capital in bad states would have negative PV.
* By sharing a controlled amount of upside ($\alpha \in [0,1]$), the investor earns the required $+10$ PV.

### Why this structure is efficient

* The firm gives up only a small fraction $\alpha$ of its high-state gains.
* In return, it avoids large destruction of value in bad states.
* The Solver solution selects the $\alpha$ that makes:

  * Debt + Equity value as large as possible, while
  * Investor PV $= +10$ (i.e., New Deal AveragePV from the firmâ€™s side is $-10$).

---

## 5. Resulting Payoff Shape

* $NT(AT)$ is **positive** in low states (investor pays the firm).
* $NT(AT)$ is **negative** in high states (firm pays investor).
* This satisfies the requirement: positive inflows for some outcomes and negative outflows for others.

The resulting payoff diagram has:

* A positive â€œspikeâ€ around the default region ($AT$ just below $F$), where the investor strongly supports the firm.
* A gently downward-sloping region for high $AT$, where the firm shares upside with the investor.

## Conclusion

The proposed New Deal:

* Eliminates or greatly reduces distress losses in bad states.
* Rewards the investor with controlled upside participation in good states.
* Maximizes the combined value of **Debt + Equity**.
* Ensures the new investor earns a **$10 present-value profit**.
* Produces a state-contingent payoff pattern (positive inflows and negative outflows) that exactly matches the assignment requirements.
